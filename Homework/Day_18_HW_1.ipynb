# 範例重點

如何使用亂數, 資料集來操作

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 設定要畫的 x, y 數據 list....
x = np.arange(0,180)
y = np.cos(x * np.pi / 180.0)
print( 'x= \n', x, '\n' )
print( 'y= \n', y )

# 畫圖
plt.plot(x,y)
plt.show() 
```

畫出完整的 sin 圖形

```python
#畫出完整的 sin 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
print( 'x= \n', x, '\n' )
print( 'y= \n', y_sin )

plt.plot(x, y_sin)

plt.show()
```

畫出完整的 cos 圖形

```python
#畫出完整的 cos 圖形
x = np.arange(0, 3 * np.pi, 0.1)
y_cos = np.cos(x)
print( 'x= \n', x, '\n' )
print( 'y= \n', y_cos )

plt.plot(x, y_cos)

plt.show()
```

**散點圖: Scatter Plots**
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)

print( 'X= \n', X, '\n' )
print( 'Y= \n', Y )
```

```python
plt.scatter( X, Y, s=75, c=np.sin(X/Y), alpha=.4 )
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
T = np.arctan2(Y,X)
plt.scatter(X,Y, s=75, c=T, alpha=.4)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
