**2. 請將 Dcard API 取得所有的看板資訊轉換成 DataFrame，並且依照熱門程度排序後存成一個 csv 的檔案。**

```python
import requests
from bs4 import BeautifulSoup
response = requests.get(
    "https://www.dcard.tw/forum/all")
soup = BeautifulSoup(response.text, "html.parser")

print(soup.prettify())
```

```python
result = soup.find_all("h3")
print(result)
```

```python
results = soup.find_all("h3")
for result in results:
    print(result.getText())
```

```python
result = soup.find_all("h4")
print(result)
```

```python
titles = soup.find_all("h4")
for title in titles:
    print(title.select_one("a").getText())
    print("https://www.dcard.tw"+title.select_one("a").get("href"))#取得文章連結
```

```python
results = soup.find_all("h3")
titles = soup.find_all("h4")
for result in results:
    print(result.getText())
    for title in titles:
        print(title.select_one("a").getText())
        print("https://www.dcard.tw"+title.select_one("a").get("href"))#取得文章連結
```

```python
resp = requests.get('https://www.dcard.tw/service/api/v2/posts')
json_resp = json.loads(resp.text)
with open('Dcard_articles.json', 'w', encoding='utf-8') as f:
    json.dump(json_resp, f, indent=2,sort_keys=True, ensure_ascii=False)
```
