```python
### 利用 PANDAS 取得酒的品質資料 
df_red = pd.read_csv("winequality_red.csv")
df_white = pd.read_csv("winequality_white.csv")
```

資料整理

```python
#資料整理
df_red["color"] = "R"
df_white["color"] = "W"
```

整合紅酒與白酒的資料

```python
#整合紅酒與白酒的資料
df_all=pd.concat([df_red,df_white],axis=0)
```

**熱圖(heatmap)**

`sns.heatmap`

`.corr()`:相關性


熱圖是數據的2D可視化,其中兩個要素之間的關係量級由色調表示。

熱圖中的**梯度根據屬性之間的相關性強度而變化。**

在下面的示例中,**高度相關的屬性的陰影比其餘屬性暗。**

```python
#Plotting heatmap
f, ax = plt.subplots(figsize=(10, 6)) #規範圖形大小
b = sns.heatmap(df_all.corr(), annot=True, linewidths=.05, ax=ax)
f.subplots_adjust(top=0.93)
bottom, top = ax.get_ylim() #設定圖形上, 底 的label
ax.set_ylim(bottom + 0.5, top - 0.5)
title= f.suptitle('Correlation Heatmap for wine attributes', fontsize=12)
```

**聯合圖**

`sns.jointplot`

聯合圖用於顯示兩個變數之間的關係。

可以選擇從 5 個選項中繪製的繪圖種類 - **散點、reg、十六進位、kde、resid。**

下面展示了三個使用線性回歸線('reg'的例子)
1. 第一個示例"檸檬酸"和"固定酸度"具有正相關關係,因此圖形是向上的。
2. 然而,硫酸鹽和酒精的變數是相關的。因此,線性回歸線幾乎是平的。
3.具有"揮發性酸度"和"檸檬酸"屬性的示例具有負相關性,因此圖呈向下。

`.reset_index(drop=True)` **concat後重排index**

```python
df_all1=df_all.reset_index(drop=True)
df_all1
```

```python
df_all1.index.is_unique
```

1. 第一個示例"檸檬酸"和"固定酸度"具有正相關關係,因此圖形是向上的。

```python
#Plotting Jointplot, 使用 'reg'== regression 回歸線繪製關係圖
sns.jointplot("fixed_acidity","citric_acid",data = df_all1,kind ='reg', color = None)
```

2. 然而,硫酸鹽和酒精的變數是相關的。因此,線性回歸線幾乎是平的。

```python
sns.jointplot("alcohol", "citric_acid", data = df_all1, kind = 'reg')
```

3.具有"揮發性酸度"和"檸檬酸"屬性的示例具有負相關性,因此圖呈向下。

```python
sns.jointplot("volatile_acidity", "citric_acid", data = df_all1, kind = 'reg')
```
