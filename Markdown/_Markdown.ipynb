假設檢定是先對母體參數提出假設（主張），
然後利用樣本的資訊，再決定是否接受或否決該假設

**如果我們能夠設法證明該說法的對立面是錯的，那就相當於證明了該說法本身是對的，**<br>
**所以建立兩個互相對立的假設**


  
虛無假設(null hypothesis) : 對母體參數提出一個主張，假設此主張為真實（除非能證明此主張非真！）

對立假設(alternative hypothesis) : 對立假設是相對於虛無假設所提出的另一個不同（相反）的假設或主張，必須有足夠的證據，才能說明此主張為真

### **假設檢定**

**如果我們能夠設法證明該說法的對立面是錯的，那就相當於證明了該說法本身是對的，**<br>
**所以建立兩個互相對立的假設**

- 對立假設(alternative hypothesis) :<br>
  用來主張**自變項的效果存在**，用符號$H_1$來表示。<br>
  換句話說，對立假設闡述研究結果裡組別之間所產生的差異是因為自變項的關係。<br>


- 虛無假設(null hypothesis) : <br>
  虛無假設正好是對立假設的相反，用來主張自變項的效果不存在，用符號$H_0$來表示。<br>
  換句話說，虛無假設闡述研究結果裡組別之間不會因為自變項而有所不同。<br>

（研究者想證明自己的研究結論是正確的，**對立假設的方向就要與想要證明其正確性的方向一致；**<br>
  同時將研究者想收集證據證明其不正確的假設作為原假設$H_0$）
  

<br>
對立假設和虛無假設彼此互斥且周延（mutually exclusive and exhaustive），<br>

**代表除了這兩個假設以外，不會有其他的假設存在，而且僅有一個假設能夠解釋最後的研究結果。**

例子:<br>
有位犯罪學者想探討守望相助隊的設立是否會影響社區的犯罪率，<br>
該研究的兩種假設的寫法分別為：<br>
對立假設（$H_1$）：守望相助隊的設立會影響社區的犯罪率。<br>
虛無假設（$H_0$）：守望相助隊的設立不會影響社區的犯罪率。

研究假設還可再區分為是否具有方向性，上面兩個假設為無方向性（non-directional）的寫法。<br>
若是有方向性（directional）的假設，則須明確地指出自變項效果的方向，以下做進一步的討論。

### **研究假設的方向性**

研究假設可為無方向性或有方向性的假設，<br>
若是**無方向性**的假設，通常會使用「會影響」、「有效果」、「有關聯」、「有所不同」等寫法

若是有方向性的假設，須在對立假設（$H_1$）中明確地**闡述自變項效果的方向，**<br>
虛無假設（$H_0$）則須主張自變項不具有對立假設（$H_1$）所闡述方向的效果（代表自變項沒有效果或具有相反於對立假設所闡述方向的效果）。

同樣以上面守望相助隊的例子來說明，<br>
若犯罪學者推測守望相助隊的設立會減少社區的犯罪率，則兩種假設的寫法分別為：<br>
對立假設（$H_1$）：守望相助隊的設立會降低社區的犯罪率。<br>
虛無假設（$H_0$）：守望相助隊的設立不會降低社區的犯罪率。

此種虛無假設的寫法帶有兩種意涵：<br>
守望相助隊的設立不會影響社區的犯罪率或是會增加社區的犯罪率。<br>
從上述的寫法也可以看出，這兩種假設是互斥且周延，**若其中一者是錯誤的，另一者必定為真實。**

### **設定顯著水準**

擬定研究假設後，須設定一個顯著水準（significance level）。<br>
顯著水準應當在開始數據收集前就設定，<br>
該顯著水準是由研究者所決定，用來和步驟4的資料適配統計模型後計算而得的機率（也就是p值）相比較的一個機率值，<br>
又稱為α水準（alpha level），通常為0.05或0.01。<br>
模型適配的機率（p-value），代表的是機率。它是反映**某一事件發生的可能性大小。**<br>
當p ⩽ α時，就可以認為結果具有統計學意義，或數據之間具有了顯著性差異。<br>

如果一件事情發生的可能性小於0.05，就可以定義為小概率事件了，<br>
也就是說，在一次研究中該事件發生的可能性很小，如果只進行一次研究，可以視為不會發生

例子:<br>
假如100次檢測中有5次或少於5次會誤判「實際上成立的虛無假設」，則以α= .05表示；<br>
設定α值愈小，表示希望檢測時的誤判機率愈低(即希望檢測能愈準確)。

由於沒有研究者能夠保證自己的研究結果百分之百的正確，<br>
所以顯著水準也可被視為研究者**願意接受第一類型錯誤的機率**，也就是承擔**拒絕一個真實的虛無假設的機率**。<br>
換句話說，若將顯著水準設為0.05，代表在只有機會的影響之下，如果p值在100次當中等於或小於5次，研究者才願意拒絕虛無假設並接受對立假設。<br>


顯著水準的選擇並沒有黃金準則，<br>
**數值愈小代表愈嚴苛的顯著水準**，也變得愈不容易拒絕虛無假設，<br>
雖然**降低第一類型錯誤的機率，卻增加了第二類型錯誤的機率**，也就是**保留一個錯誤的虛無假設**。

假設檢定用來評估自變項是否有效果，並決定是否拒絕虛無假設。<br>
但評估虛無假設是否為真實的過程中，研究者可能會犯下兩種錯誤：第一類型錯誤和第二類型錯誤。<br>

### **決策規則**

研究假設分為虛無假設和對立假設，兩者互斥且周延，當其中一者為錯誤，另一者必定為真實，且沒有其他的可能性存在。<br>
研究者拒絕或保留虛無假設的決策規則如下：

若數據結果的機率 $\leq \alpha$，拒絕虛無假設。<br>
若數據結果的機率 $> \alpha$，保留虛無假設。<br>

當研究者拒絕虛無假設，接受對立假設時，此種研究結果稱為顯著（significant）；<br>
反之，若無法拒絕虛無假設，也就是必須保留虛無假設時，研究結果則稱為不顯著（non-significant）。<br>


|$\alpha$|p-value|研究結果|
|---|---|---|
|0.05|0.03|拒絕虛無假設|
|0.01|0.03|保留虛無假設|

從決策規則可看出顯著水準的選擇會影響研究的結果。<br>
若數據結果的機率為0.03，當顯著水準設為0.05時，因為0.03<0.05，所以拒絕虛無假設，接受對立假設；<br>
但當顯著水準改設為0.01時，因為0.03>0.01，則無法拒絕虛無假設。<br>
由此可知，0.01是一個比0.05更為嚴苛的顯著水準，因為研究者較不容易拒絕虛無假設。<br>

顯著水準的選擇並沒有黃金準則，通常會依據研究的性質、目的和錯誤結論可能造成的後果來決定。<br>
若一個研究為**探索（exploratory）性質**，目的在於初步瞭解一個議題，研究完成後再決定是否進行更深入的探討，可選擇**較寬鬆的顯著水準**；<br>
若一個研究為**解釋（explanatory）性質**，目的在於建立因果關係並推論至其他群體，因為錯誤的結論可能造成嚴重的後果，所以需要**較嚴格的顯著水準。**

**選擇顯著水準時還須注意研究假設是否具有方向性**<br>
沒有方向性的假設檢定稱為雙尾檢定（two-tailed test）<br>
有方向性的假設檢定則稱為單尾檢定（one-tailed test）<br>

### **第一類型錯誤和第二類型錯誤的意義**

假設檢定的結果不外乎兩種：<br>

自變項有效果，所以拒絕虛無假設，接受對立假設；<br>
或自變項沒有效果，所以保留虛無假設。<br>

不過，研究者無法百分之百地確定研究結果一定是正確的，<br>
**很有可能自變項實際上沒有效果，但研究結果顯示有效果；**<br>
**或自變項實際上有效果，但研究結果卻指出沒有效果。**<br>
因此，假設檢定的決策可能會出現兩種錯誤：<br>

第一類型錯誤（Type Ⅰ error）：拒絕一個真實的虛無假設。<br>
第二類型錯誤（Type Ⅱ error）：保留一個錯誤的虛無假設。<br>

第一類型的錯誤是指**研究者相信自變項帶有效果，但實際上自變項並沒有效果**。<br>
該錯誤的機率通常為0.05，即是資料蒐集前已設定好的顯著水準，也稱為α水準（α-level）。<br>
換句話說，若複製研究100次，研究者可預期其中有5次會出現拒絕虛無假設的結果，但實際上自變項並沒有效果。<br>

第二類型錯誤是指**研究者相信自變項沒有效果，但實際上自變項帶有效果**。<br>
Cohen（1992）建議能夠被接受的第二類型錯誤的最大機率為0.2，稱為β。<br>
也就是說，若從自變項帶有效果的母群體中，隨機抽取出樣本100次，則其中會有20次研究者無法測量到自變項的效果。<br>

檢定過程的最後，研究者透過決策規則，所得到的研究結果和母群體的真實狀況之間會有下面表格所呈現的4種結果：



|||**真實的狀況**|
|---|---|---|
|**研究結果**|*虛無假設是真實的*|*虛無假設是錯誤的*|
|*保留虛無假設*|正確決定|第二類型錯誤|
|*拒絕虛無假設*|第一類型錯誤|正確決定|

當保留真實的虛無假設或拒絕錯誤的虛無假設時，皆為正確的決定。<br>
不過，當研究者拒絕虛無假設，但實際上虛無假設是真實的時候，研究者犯了第一類型錯誤。<br>
反之，當研究者保留虛無假設，但實際上虛無假設是錯誤的時候，研究者則犯了第二類型錯誤。<br>

但是，第一類型和第二類型錯誤**並非各自獨立**，兩者之間實具有密切的關係。<br>
**研究者可能採用較為嚴苛的顯著水準，試圖降低第一類型錯誤的機率，卻因此提升了第二類型錯誤的風險**。<br>


### **第一類型錯誤和第二類型錯誤間的關聯**

若研究者想降低第一類型錯誤的機率，選擇將顯著水準設為0.01，由於$p=0.03> \alpha =0.01$，所以保留虛無假設。<br>
然而，設定較嚴苛的顯著水準，雖然降低了拒絕真實的虛無假設之機率，卻提高了保留錯誤的虛無假設（第二類型錯誤）之風險

|顯著水準$\alpha$|模型適配的機率（p-value）|||**真實的狀況**|
|---|---|---|---|---|
|||**研究結果**|*虛無假設是真實的*|*虛無假設是錯誤的*|
|0.01|0.03|*保留虛無假設*|正確決定|第二類型錯誤|
|0.05|0.03|*拒絕虛無假設*|第一類型錯誤|正確決定|


由於第一類型錯誤和第二類型錯誤是立基於不同的假設：<br>
犯下第一類型的錯誤必須是母群體中**自變項不存在效果**，<br>
第二類型的錯誤則是母群體中**自變項存在效果**，所以兩種錯誤之間的關係並非顯而易見。<br>
即便知道降低第一類型錯誤的機率會提升第二類型錯誤的機率，也很難訂立一標準去取得兩種錯誤之間的平衡。<br>

雖然第一類型錯誤和顯著水準有關聯，但這並不表示真相取決於顯著水準的數值，<br>
嚴苛的顯著水準只是降低犯下第一類型錯誤的機率而已。<br>
**顯著水準的選擇通常會因為研究的性質、目的和錯誤的結論可能造成的後果而有不同。**<br>

若研究為**探索（exploratory）性質**，目的在於初步探究一個議題，視研究結果再決定是否進行更深入的探討，<br>
**第一類型的錯誤通常可被容忍**，所以可選擇**較寬鬆的顯著水準**。<br>
反之，若研究為**解釋（explanatory）性質**，目的在於建立因果關係並推論至其他的群體，<br>
因為錯誤的結論可能帶來嚴重的後果，所以研究者寧願**降低第一類型錯誤的機率，承擔第二類型錯誤的風險**。<br>


由於沒有一個研究能夠確信研究結果為正確的，所以需要**重複驗證（replication）**。<br>
在一個研究結果能夠廣泛地被科學界接受前，必須經過反覆的驗證，而非單一研究的發現即可被視為真實。<br>
因此，重複驗證不但能夠提高研究結果的可信度，也可以大幅降低犯下第一類型錯誤的機率。<br>

### **假設檢定的種類**

**選擇顯著水準時還須注意研究假設是否具有方向性**<br>
沒有方向性的假設檢定稱為雙尾檢定（two-tailed test）<br>
有方向性的假設檢定則稱為單尾檢定（one-tailed test）<br>


若是無方向性，通常會使用「會影響」、「有效果」或「有關聯」等寫法，例如興建社區活動中心**會影響**社區裡少年的犯罪率。<br>
若是有方向性的假設，則須在對立假設中**明確地寫出自變項效果的方向**，例如興建社區活動中心**會降低**社區裡少年的犯罪率。<br>

**雙尾和單尾檢定**

雙尾檢定（two-tailed test）:<br>
$\rightarrow$沒有方向性<br>
由於沒有方向性的研究假設不預測自變項效果的方向，所以**顯著水準會平均地劃分在分配的兩側**，<br>
若以0.05的顯著水準為例，一側各為0.025，如下圖左右兩側尾端的深藍色區域。<br>
依據上述的決策規則，研究數據結果**在單側的機率必須等於或小於0.025，才可以拒絕虛無假設並接受對立假設**。<br>
因為顯著水準會平均劃分在分配的兩端，所以無方向性的假設檢定稱為雙尾檢定。<br>

單尾檢定（one-tailed test）:<br>
$\rightarrow$有方向性<br>
若是有方向性的研究假設，依據假設所預測的方向，**顯著水準會集中在分配的一側**，<br>
例如下圖深藍色加淺藍色的區域。<br>
根據決策規則，**研究數據結果的機率等於或小於0.5且與預測的方向相同時，即可拒絕虛無假設**。<br>
由於顯著水準集中在一側，此種有方向性的假設檢定稱為單尾檢定。<br>

![Image of 1_and_2-tailed_tests](https://drfishstats.com/wp-content/uploads/2021/11/1_and_2-tailed_tests.jpg)

若單尾檢定的研究結果顯示自變項的效果剛好跟預測的方向相反，由於此種情況無法拒絕虛無假設，所以可能錯失發現自變項效果的機會。

原則上，除非有下列的兩種情況，較適合使用單尾檢定外，**建議所有的研究都儘量採取雙尾檢定的假設檢定方式**：

1. **即使研究結果和預測的方向完全相反，也會被視為沒有實際上的差別或助益，僅能導致保留虛無假設的結果**（Ruxton & Neuhäuser，2010）。<br>
   例如對立假設闡述興建社區活動中心會降低社區裡少年的犯罪率，但研究結果卻顯示社區活動中心的興建增加少年的犯罪率；<br>
   由於興建社區活動中心非但沒有改善犯罪反而導致犯罪的增加，便沒有核准該項公共設施興建的意義存在。<br>

2. **有很強的理論基礎，且有相關的數據佐證，證明單尾檢定是適當的假設檢定方式。**<br>
   若研究結果和預測的方向完全相反，雖然最後還是只能保留虛無假設，但卻可藉此檢視理論與先前的研究發現。<br>

補:<br>
有些研究者可能在看到研究結果後產生修改研究假設的衝動，<br>
例如單尾檢定改成雙尾檢定或雙尾檢定改成單尾檢定，但此種作法不僅違反研究程序，還會增加研究錯誤的機率。<br>
因此，研究者不但需要在資料蒐集前設定好顯著水準，且決定好採用雙尾或單尾檢定，更需要將這些設定貫徹至研究結束，<br>
即便研究結果和預測的方向完全相反，也應該虛心接受，並保留虛無假設。<br>

## **選擇合適的統計檢定方式**

闡述完研究假設和設定好顯著水準後，須選擇合適的統計檢定方法。

統計檢定方法有很多種，依據**研究設計、資料分布狀態、變項的測量尺度**而有所不同。


### **母數分析方法 vs. 無母數分析方法**

一般而言，統計檢定可分為母數檢定（parametric test）和無母數檢定（non-parametric test）。

有母數統計資料是根據從中**採取樣本之母體分佈的假設。**<br>
無母數統計資料不是根據假設，亦即，**可以從不遵循特定分佈的樣本中收集資料**<br>

若資料呈現**常態分布**，**母數檢定**為合適的統計檢定方法；

若資料沒有呈現常態分布，無母數檢定則較為合適。

- 母數分析方法(檢定)：<br>
  一般假設母群體為**常態分布**以進行檢定的方式，**Pearson相關係數、t 檢定、變異數分析、線性迴歸**等皆屬之。<br>


- 無母數分析方法(檢定)：<br>
  無法符合有母數分析所設計的方法；<br>
  常使用**符號(正負)**或**排序(大小順序)取代測量數值**，或使用**各分類的次數**以進行統計分析。<br>
  適用於類別、序位尺度資料分析與資料分布未知的情況。
  
  - 優點：<br>
    母群體**分布未知或不是常態分布**，或是**樣本數不夠大**時皆可使用。是無母數分析方法的最大優點。<br>
    計算簡單且快速。<br>
    雖然在母群實際上為常態分配時，較有母數分析方法不易得到顯著結果；<br>
    但**在母群體不是常態分布時，無母數分析方法之檢力較有母數分析方法高**。<br><br>
    
  - 缺點：<br>
    只使用**資料的符號、排序**等特性，浪費了數值之集中趨勢、分散性及分佈所提供的資訊。<br>
    **針對常態分布資料如果仍進行無母數分析，將使檢力降低**。<br>
  
當欲檢定的資料不符合有母數分析法之假設前提時才建議使用無母數分析法，<br>
為一種**互補的統計方法，而非用於取代有母數分析法**。<br>

**平均數抽樣分布**:<br>

是指從母群體中隨機抽取出樣本大小為$N$的所有可能樣本，<br>
計算出每一個樣本的平均數與獲得該平均數的機率，<br>
最後呈現出各個平均數機率分布情形的機率分配。<br>

平均數抽樣分配有3個特性：

1. $\mu_{\overline X}=\mu$：平均數抽樣分配的平均數等於母群體平均數。

2. $\sigma_{\overline X}=\frac {\sigma}{\sqrt N}$ ：平均數抽樣分配的標準差（標準誤）等於母群體標準差除以$\sqrt N$

3. 中央極限定理：不管母群體本身的分布型態，只要樣本數愈大，平均數抽樣分配就會愈趨近於常態分配。

平均數抽樣分配的定義和特性 : https://drfishstats.com/inferential-stats/sampling-distribution-of-mean/

### **單一樣本Z檢定的假設檢定** :<br>

單一樣本Z檢定用來**比較一個樣本平均數和母群體平均數間是否有顯著的差異**<br>

單一樣本Z檢定的基本統計量為**樣本平均數**，用來和一個已知的母群體平均數比較，<br>
須透過**平均數抽樣分配（sampling distribution of the mean）來計算獲得特定研究結果的機率值（p值），才能評估拒絕或保留虛無假設**。<br>

計算Z檢定統計量的公式和求得Z分數（也就是標準分數）的公式非常相似，唯一的差別在於:<br>
Z分數的計算使用**原始數據**<br>
Z檢定統計量的計算則使用**平均數**<br>

![Image of z-score_vs_z-statistic](https://drfishstats.com/wp-content/uploads/2022/01/z-score_vs_z-statistic.jpg)


根據上面提到的平均數抽樣分配的前2個特性，Z檢定統計量的計算公式可變成：

$z=\frac {\overline X-\mu_{\overline X}}{\sigma_{\overline X}}=\frac {\overline X-\mu}{\displaystyle \frac {\sigma}{\sqrt N}}$

由於平均數抽樣分配在樣本數夠大時會呈現常態分配，<br>
所以計算出Z檢定統計量後，可藉由標準常態分配表查詢得到該統計量的機率，<br>
若該機率等於或小於事先設定好的顯著水準，即可拒絕虛無假設，接受對立假設。<br>

單一樣本Z檢定用來比較一個樣本平均數和母群體平均數間是否有顯著的差異，其假設檢定的過程並不困難，但在使用該檢定前，資料須滿足以下3個條件：

1. 欲評估的基本統計量為樣本平均數（$\overline X$）<br>
2. 母群體的平均數$\mu$和標準差$\sigma$為已知的情況。<br>
3. 平均數抽樣分配呈現常態分配，也就是$N \geq 30$，或母群體本身呈現常態分配。<br>

評估研究結果的決策規則有兩種，<br>
- 比較p值和α水準<br>
  計算Z檢定統計量，查標準常態分配表，得到該統計量的機率 B(超過標準分數 z 之外的面積)，作為p值，並跟α水準比較<br>
  **若p<α**，即可拒絕虛無假設，接受對立假設<br>
  (計算得到Z檢定統計量的數值後，即可查詢標準常態分配表，找到獲得該統計量的機率（p值），再運用決策規則，比較p值和事先選擇好的顯著水準（α水準），評估是否拒絕虛無假設。)
  


- 比較檢定統計量和臨界值<br>
  先查α水準(0.05)在標準常態分配表的Z score，α水準是B(超過標準分數 z 之外的面積)，他的Z score介於1.64~1.65<br>
  比較Z檢定統計量跟α水準的Z score<br>
  **若Z檢定統計量>α水準的Z score**，代表該值落在臨界區域，所以可拒絕虛無假設，接受對立假設。<br>
  (藉由事先選擇好的α水準，查詢標準常態分配表，找出相對應的臨界值，並比較**z統計量和臨界值**。**若z統計量的絕對值*等於或大於臨界值*的絕對值，即可拒絕虛無假設，接受對立假設；反之，則須保留虛無假設。**)

### **獨立樣本z檢定的概念**

獨立樣本z檢定（independent samples z-test）用來**比較兩個不同群體的平均數是否有顯著差異**

獨立樣本z檢定須在**兩群體的母群體變異數皆已知**的情況下才可使用，所以**實際上幾乎不會用到**。

雖然實際研究過程不太可能使用到獨立樣本Z檢定，但該種檢定方法卻是瞭解獨立樣本t檢定的一個重要概念

**獨立樣本的研究設計**

獨立樣本的研究設計是指研究包含了兩種以上的群體，<br>
而這些群體可能為原本即存在的團體或實驗分配的團體，<br>
分別稱為**原樣團體（intact group）設計**和**真實驗（true experiment）設計**，下表為兩種設計的比較。

![Image of types_of_independent_groups_design](https://drfishstats.com/wp-content/uploads/2022/03/types_of_independent_groups_design.jpg)

原樣團體和真實驗設計的差別:<br>
在於**自變項（independent variable）的本質或種類**，<br>
而自變項是研究人員用來**描述或解釋研究結果的一種變項**，<br>
又可區分為受試者變項（subject variable）和操縱變項（manipulated variable）。

- 受試者變項:<br>
  是指研究人員無法操縱的情境，例如研究參與者的生理性別、年齡、智商或經濟收入。<br>
  運用受試者變項的獨立樣本設計**可以探討群體之間是否有差異，也可以推估差異的程度，但無法做出因果關係的推論**。<br>


- 操縱變項:<br>
  是指研究人員可以直接操縱的情境，而且**可隨機分配研究參與者至任一情境，例如實驗組和對照組**。<br>
  運用操縱變項的獨立樣本設計，當研究人員隨機分配研究參與者至不同的情境時，即是創造出不同的假設母體（hypothetical population）。<br>
  若研究結果指出不同情境之間的差異相當大，研究人員就可假設樣本來自的不同母群體之間也存在著差異，因此可做出操縱變項導致了該差異的結論。<br>
  
基本上，獨立樣本的研究設計會**包含至少兩種情境，例如女性和男性、實驗組和對照組**。<br>
進行**兩種情境**的資料分析時，可使用**獨立樣本Z檢定或獨立樣本t檢定**；<br>
若是**超過兩種情境**的資料分析，就須使用較為複雜的**變異數分析（analysis of variance）**。<br>


相較於單一樣本的研究設計（單一樣本z檢定、單一樣本t檢定），獨立樣本的研究設計有下列3點的優勢：<br>

1. **無須知道母群體平均數μ**：<br>
   單一樣本的研究設計須明確指出母群體的平均數，才能用來和樣本平均數比較，但大多數的時候，**母群體平均數的資訊很難取得**；<br>
   相對地，獨立樣本的研究設計無須知道母群體平均數即可進行假設檢定。


2. **無須考量母群體平均數μ的正確性**：<br>
   如上一點所言，單一樣本的研究設計須使用母群體平均數才能進行假設檢定，<br>
   且該數值的正確性可能存有疑義，例如該數值可能為好幾年前的資訊、數值的取得過程可能不客觀。<br>
   相對地，獨立樣本的研究設計無須使用到母群體平均數，所以不會遇到母群體平均數正確與否的問題。<br>


3. **可做因果關係的推論**：<br>
   單一樣本的研究設計在於比較一個樣本平均數是否顯著地不同於一個母群體平均數，<br>
   研究結果僅能顯示兩者是否存在差異，無法做出任何因果關係的推論。<br>
   相對地，若是使用**真實驗的獨立樣本研究設計**，將研究參與者隨機分配至不同的情境，<br>
   可藉由**不同情境之間的差異程度導出操縱變項是否造成該差異的結論。**
   
**獨立樣本的研究假設**

假設在只有**兩個情境**的獨立樣本裡，<br>
第1個樣本的分數是來自於第1個**常態分配母群體的一組隨機樣本**，<br>
將第1個假設母體的平均數符號寫為$\mu_1$、標準差符號寫為$\sigma_1$。<br>
第2個樣本的分數是來自於第2個**常態分配母群體裡的一組隨機樣本**，<br>
將第2個假設母體的平均數符號寫為$\mu_2$、標準差符號寫為$\sigma_2$。<br>


  

研究假設:<br>
可分為對立假設（$H_1$）和虛無假設（$H_0$）<br>
依據方向:<br>
劃分為無方向性和有方向性的假設<br>

- 無方向性的研究假設<br>
  - 對立假設（$H_1$）：第1個樣本的平均數不同於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 \neq \mu_2$的隨機樣本。<br>
  - 虛無假設（$H_0$）：第1個樣本的平均數和第2個樣本的平均數相同，也就是樣本是來自於母群體$\mu_1 = \mu_2$的隨機樣本。<br>


- 有方向性的研究假設<br>

  若明確地指出自變項效果或作用的方向，則為有方向性的研究假設，可能會有$\mu_1 > \mu_2$或$\mu_1 < \mu_2$的兩種情況，須分開來看。<br>

  - 若$\mu_1 > \mu_2$<br>
    - 對立假設（$H_1$）：第1個樣本的平均數大於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 > \mu_2$的隨機樣本。<br>
    - 虛無假設（$H_0$）：第1個樣本的平均數沒有大於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 \leq \mu_2$的隨機樣本。<br>
 
  - 若$\mu_1 < \mu_2$<br>
    - 對立假設（$H_1$）：第1個樣本的平均數小於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 < \mu_2$的隨機樣本。<br>
    - 虛無假設（$H_0$）：第1個樣本的平均數沒有小於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 \geq \mu_2$的隨機樣本。<br>
  
! 獨立樣本的假設檢定有一個前提假設，<br>
  即第1個母群體的變異和第2個母群體的**變異相同**，也就是$\sigma_1^2 = \sigma_2^2$，稱為**變異數同質性（homogeneity of variance）假設**。<br>
  當自變項確實有效果或作用時，該效果會影響到群體裡的每一個人，因此**只會改變母群體的平均數，但不會改變母群體的標準差或變異**。<br>
  
**樣本平均數差異的抽樣分配**<br>

兩個情境的獨立樣本的假設檢定使用樣本平均數差異的抽樣分配（sampling distribution of the difference between sample means），<br>
是指從平均數為$\mu_1$、變異數為$\sigma_1^2$的第1個母群體中隨機抽取出樣本大小為$n_1$的所有可能的樣本；<br>
同樣地，從平均數為$\mu_2$、變異數為$\sigma_2^2$的第2個母群體中隨機抽取出樣本大小為$n_2$的所有可能的樣本。<br>

抽取出所有可能的樣本後，<br>
計算出每一個樣本的平均數，$\overline X_1$和$\overline X_2$，<br>
再計算出所有可能配對的$\overline X_1-\overline X_2$的數值以及獲得各個數值的機率，<br>
藉此獲得的機率分配型態即為樣本大小為$n_1$、$n_2$的樣本平均數差異的抽樣分配。<br>

重複上述的過程至不同的樣本大小和母群體分數，就能建構出樣本平均數差異的抽樣分配。

樣本平均數差異的抽樣分配有以下幾個特徵：
- 若樣本來自的母群體呈現常態分配，樣本平均數差異的抽樣分配也會呈現常態分配。<br>
  在樣本大小為$n_1$和$n_2$的情況下，從$\overline X_1-\overline X_2$所有可能配對的數值，<br>
  可以求出這些數值的平均數和標準差，<br>
  也就是樣本平均數差異抽樣分配的平均數$\mu_{\overline X_1-\overline X_2}$和標準差$\sigma_{\overline X_1-\overline X_2}$<br>
  

- $\mu_{\overline X_1-\overline X_2}=\mu_1-\mu_2$：
  樣本平均數差異抽樣分配的平均數等於兩個母群體平均數的差值。<br>
  

- $\sigma_{\overline X_1-\overline X_2}=\sqrt {\sigma_{\overline X_1}^2+\sigma_{\overline X_2}^2}$：<br>
  只要是從母群體中抽取出樣本，就一定會存在變異。<br>
  樣本平均數差異抽樣分配的標準差，還可以稱為樣本平均數差異的標準誤（standard error of the difference between sample means），<br>
  等於樣本大小為$n_1$的平均數抽樣分配之變異數加上樣本大小為$n_2$的平均數抽樣分配之變異數後開根號。
  
  ![Image of types_of_graph_sampling_distribution_of_the_difference_between_sample_means](https://drfishstats.com/wp-content/uploads/2022/03/graph_sampling_distribution_of_the_difference_between_sample_means.jpg)
  
由於樣本平均數差異的抽樣分配呈現常態分配，所以獨立樣本Z檢定可以使用Z檢定統計量和Z分配（常態分配）

**獨立樣本Z檢定的使用**

獨立樣本Z檢定是用來比較**兩個樣本的平均數**是否有顯著不同的一種統計檢定方法，而兩個樣本皆須來自於**常態分配的母群體**。<br>
進行假設檢定時，須使用**樣本平均數差異的抽樣分配**。<br>

從樣本平均數差異的抽樣分配的特徵得知，<br>
**若兩個樣本來自的母群體呈現常態分配，樣本平均數差異的抽樣分配也會呈現常態分配，**<br>
所以獨立樣本ｚ檢定的假設檢定使用Z檢定統計量和常態分配。

獨立樣本Z檢定的檢定統計量的計算公式如下：

$\begin{equation*}z = \frac {(\overline X_1-\overline X_2)-\mu_{\overline X_1-\overline X_2}}{\sigma_{\overline X_1-\overline X_2}}\end{equation*}$

將上面提到的樣本平均數差異的抽樣分配之特徵帶入該公式中，Z檢定統計量的公式會變成：

$\begin{equation*}z = \frac {(\overline X_1-\overline X_2)-(\mu_1-\mu_2)}{\sqrt {\sigma_{\overline X_1}^2+\sigma_{\overline X_2}^2}}\end{equation*}$

計算得到Z檢定統計量的數值後，<br>
即可查詢**標準常態分配表**，找到獲得該統計量的機率（p值），<br>
再運用決策規則，比較p值和事先選擇好的顯著水準（$\alpha$水準），評估是否拒絕虛無假設。<br>

然而，從上面的公式中可以看出，<br>
兩個母群體的變異數必須是已知的情況下，才可以計算出Z檢定統計量。<br>
由於大多數的時候，母群體變異數難以取得，<br>
所以**在比較兩個平均數的獨立樣本研究裡，幾乎不會用到獨立樣本Z檢定，而是使用樣本統計量推估母體參數的獨立樣本t檢定。**<br>

### **單一樣本t檢定的假設檢定**

單一樣本t檢定和單一樣本Z檢定的比較

單一樣本t檢定和單一樣本Z檢定同樣用來比較**一個樣本平均數和一個已知的母群體平均數，**<br>
但不同於**單一樣本Z檢定**是在**母群體標準差已知**的情況下使用，**單一樣本t檢定**是在**母群體標準差未知**的情況下使用。<br>
*因為母群體標準差未知，所以單一樣本t檢定用**樣本標準差估計母群體標準差**。*<br>


單一樣本t檢定用來比較一個樣本平均數和一個已知的母群體平均數，但在使用前資料須滿足下述的兩個條件：

1. 基本統計量為**一個樣本平均數（$\overline X$），且母群體平均數$\mu$已知**，但**母群體標準差$\sigma$未知**。
2. 平均數抽樣分配須呈現常態分配，也就是樣本數等於或大於30，或樣本來自的母群體呈現常態分配。


下表為兩者公式的比較：

![Image of types_of_formulas_one-sample_z-test_vs_t-test](https://drfishstats.com/wp-content/uploads/2022/01/formulas_one-sample_z-test_vs_t-test.jpg)

從上表可看出，兩個公式的分子並無不同，差別在於**分母的計算**。<br>
單一樣本Z檢定公式的分母為**母群體標準差$\sigma$除以$\sqrt N$**，而單一樣本t檢定公式的分母則為**樣本標準差s除以$\sqrt N$**。<br>
也就是說，單一樣本t檢定用估計平均數標準誤$s_{\overline X}$取代平均數標準誤$\sigma_{\overline X}$<br>

另外一個差別在於**樣本的大小**。使用單一樣本**Z檢定時，樣本數須等於或大於30**，然而單一樣本**t檢定則可使用在樣本數小於30的情況。**<br>

簡單來說，<br>
**當$N \geq 30$且母群體標準差$\sigma$已知時，使用單一樣本Z檢定；**<br>
**但當N < 30或母群體標準差$\sigma$未知時，則須使用單一樣本t檢定。**<br>

**t抽樣分配:**<br>
t抽樣分配（sampling distribution of t）是指從母群體中隨機抽取出樣本大小為N的所有可能樣本，<br>
計算出每一個樣本的t值和獲得該t值的機率，最後呈現出各個t值的機率分布狀態之機率分配。<br>

**曲線**<br>
Z分配僅有一條曲線（常態曲線）<br>
t分配會**隨著樣本大小**而有不同的曲線。**當樣本數等於或大於30時，t分配就會趨近於常態分配**。<br>
但更正確地說，t分配是隨著**自由度（degrees of freedom，簡寫為df）**的改變而有不同的曲線，而非單純地因為樣本大小而改變。<br>

**自由度（degrees of freedom，簡寫為df）**<br>
自由度是指計算一個統計量時，**可以自由變動的分數個數**。<br>
舉例來說，若隨機從一母群體裡抽取出5個分數，這5個分數都可自由地變動。<br>
但若已知平均數為15，則不可能5個分數皆可自由地變動，<br>
假設前4個分數為5、10、15、20，最後一個分數必定為25，才能使平均數為15，因此自由度為4（$N-1=5-1=4$）。<br>


t分配在不同自由度下的曲線型態。<br>
**t分配為對稱的鐘形曲線，且平均數為0，**可用來決定得到一個特定數值的機率值。<br>
**自由度愈高時，曲線會愈接近常態分配**的曲線型態。<br>
當**自由度等於30時，t分配已趨近於常態分配**；<br>
當**自由度無限大時，t分配和常態分配已全然相同。**<br>

**單一樣本t檢定的假設檢定**

單一樣本t檢定用來比較**一個樣本平均數（$\overline X$）是否顯著地不同於一個已知的母群體平均數$\mu$**，且**母群體標準差$\sigma$未知**的情況。

單一樣本t檢定的檢定統計量為t統計量，分子為**樣本平均數和母群體平均數**的差值，分母為**樣本標準差除以$\sqrt N$**<br>
公式:<br>
$ t=\frac {\overline X-\mu}{s_{\overline X}} =\frac {\overline X-\mu}{\displaystyle \frac {s}{\sqrt N}} $<br>

**決策規則:**

- 藉由事先選擇好的α水準，**查詢t分配表**，**自由度為$N-1$**，找出相對應的臨界值，並比較**t統計量和臨界值**。<br>
  **若t統計量的絕對值*等於或大於臨界值*的絕對值，即可拒絕虛無假設，接受對立假設；反之，則須保留虛無假設。**<br>


- 若使用統計分析軟體來執行單一樣本t檢定，通常會輸出獲得該特定**t統計量的機率值（p值）**，反而省卻了查詢表格的麻煩。<br>
  **若p值小於或等於α水準（$p \leq \alpha$）**，即可拒絕虛無假設，接受對立假設；反之，則須保留虛無假設。<br>
  
若使用有方向性的研究假設，須注意**t統計量和對立假設的方向要一致**，否則就應保留虛無假設。<br>
例如:<br>
對立假設為樣本平均數小於母群體平均數，計算出來的t統計量須為負數，才能進一步比較t統計量和臨界值；<br>
反之，若t統計量為正數，因為和對立假設的方向相反，所以保留虛無假設。<br>

單一樣本t檢定用來比較一個樣本平均數和一個已知的母群體平均數，但在使用前資料須滿足下述的兩個條件：

1. 基本統計量為一個樣本平均數（$\overline X$），且母群體平均數$\mu$已知，但母群體標準差$\sigma$未知。
2. 平均數抽樣分配須呈現常態分配，也就是樣本數等於或大於30，或樣本來自的母群體呈現常態分配。

單一樣本t檢定 : https://drfishstats.com/inferential-stats/statistical-tests/comparison-tests/one-sample-t-test/

### **獨立樣本t檢定的假設檢定**

獨立樣本t檢定（independent samples t-test）和獨立樣本Z檢定皆是用來比較**兩個群體的平均數**是否有顯著不同的統計檢定方法，<br>
但**獨立樣本Z檢定**須在兩群體來自的**母群體之變異數已知**的情況下才可使用。<br>
相對地，**獨立樣本t檢定**則不受此限制，因其直接**使用樣本統計量推估母體參數**來進行運算，實務上是很常被使用的一種統計檢定方法。<br>

P.S. 變異數開根號後得「標準差」

**平均數抽樣分配**

- $\mu_{\overline X}=\mu$：平均數抽樣分配的平均數等於母群體平均數。
- $\sigma_{\overline X}=\frac {\sigma}{\sqrt N}$ ：平均數抽樣分配的標準差（標準誤）等於母群體標準差除以$\sqrt N$
- $\mu_{\overline X_1-\overline X_2}=\mu_1-\mu_2$：樣本平均數差異抽樣分配的平均數等於兩個母群體平均數的差值。<br>
- $\sigma_{\overline X_1-\overline X_2}=\sqrt {\sigma_{\overline X_1}^2+\sigma_{\overline X_2}^2}$：<br>
  只要是從母群體中抽取出樣本，就一定會存在變異。<br>
  樣本平均數差異抽樣分配的標準差，還可以稱為樣本平均數差異的標準誤（standard error of the difference between sample means），<br>
  等於樣本大小為$n_1$的平均數抽樣分配之變異數加上樣本大小為$n_2$的平均數抽樣分配之變異數後開根號。

**獨立樣本t檢定和獨立樣本Z檢定的比較**

在比較一個樣本平均數和一個母群體平均數是否有顯著的不同時，**若母群體*標準差*未知，會使用樣本標準差來估計母群體標準差**，<br>
也就是用單一樣本t檢定而非單一樣本z檢定的假設檢定來進行假設檢定。<br>
同樣地，當進行**兩個樣本*平均數*的比較，且母群體*變異數*未知時，就須改用*樣本變異數*來推估母群體變異數。**<br>

根據**變異數總和法則（variance sum law）**，<br>
「兩個獨立的變項之變異等於兩個變項的變異數之和」<br>
在獨立樣本設計裡，兩個樣本平均數間彼此獨立，所以**樣本平均數差異抽樣分配的變異等於兩樣本來自的母群體變異數之總和**：<br>
$ \sigma_{\overline X_1-\overline X_2}^2=\sigma_{\overline X_1}^2+\sigma_{\overline X_2}^2 $

代入: $\sigma_{\overline X}=\frac {\sigma}{\sqrt N}$

=> $ \sigma_{\overline X_1-\overline X_2}^2=\sigma_{\overline X_1}^2+\sigma_{\overline X_2}^2=\frac {\sigma_1^2}{n_1}+\frac {\sigma_2^2}{n_2} $

運用相同的邏輯至母群體變異數未知的兩個樣本平均數比較上，利用樣本變異數$s^2$來估計母群體變異數$\sigma^2$，則樣本平均數差異的標準誤估計值為：

$\begin{equation*}s_{\overline X_1-\overline X_2}=\sqrt {s_{\overline X_1}^2+s_{\overline X_2}^2}=\sqrt {\frac {s_1^2}{n_1}+\frac {s_2^2}{n_2}}\end{equation*}$

![Image of types_of_independent_samples_equation_z-test_vs_t-test](https://drfishstats.com/wp-content/uploads/2022/03/independent_samples_equation_z-test_vs_t-test.jpg)

不過$\begin{equation*}s_{\overline X_1-\overline X_2}=\sqrt {s_{\overline X_1}^2+s_{\overline X_2}^2}=\sqrt {\frac {s_1^2}{n_1}+\frac {s_2^2}{n_2}}\end{equation*}$<br>
存在一個問題，當兩個樣本的樣本大小不一致的時候（$n_1 \neq n_2$），該公式就無法提供正確的估計值。<br>

為了解決此問題，須改用**合併變異數估計值（pooled variance estimate）**，<br>
它考量了**樣本的大小並運用自由度（degrees of freedom）加權兩個樣本的變異數**。<br>
如果將合併變異數估計值簡寫為$s_p^2$，則公式為：<br>

$\begin{align*}s_p^2 &= \frac {(n_1-1)s_1^2+(n_2-1)s_2^2}{n_1+n_2-2} \\&= \frac {(n_1-1)(\displaystyle \frac {SS_1}{n_1-1})+(n_2-1)(\displaystyle \frac {SS_2}{n_2-1})}{n_1+n_2-2} \\&= \frac {SS_1+SS_2}{n_1+n_2-2}\end{align*}$<br>

$s_1^2$和$s_2^2$分別指第1個和第2個樣本的變異數。<br>
SS稱為平方和（sum of squares），指樣本中的每個分數和樣本平均數間的差值平方後相加而得到的數值。<br>
$n_1+n_2-2$為自由度，因為每計算一個樣本標準差會失去1個自由度，這裡要計算兩個樣本的標準差，所以會失去2個自由度。<br>

因此，若從變異數未知的第1個母群體中隨機抽取出樣本大小為$n_1$的樣本，也從變異數未知的第2個母群體中隨機抽取出樣本大小為$n_2$的樣本，<br>
利用$s_p^2$，下面的公式可提供一個**無偏誤的樣本平均數差異標準誤的估計值：**

$\begin{equation*}s_{\overline X_1-\overline X_2} = \sqrt {s_p^2(\frac {1}{n_1}+\frac {1}{n_2})} = \sqrt {(\frac {SS_1+SS_2}{n_1+n_2-2})(\frac {1}{n_1}+\frac {1}{n_2})}\end{equation*}$<br>

得到**樣本大小不一致時可以使用的獨立樣本t檢定的t檢定統計量公式：**<br>
=> $\begin{equation*}t=\frac {(\overline X_1-\overline X_2)-(\mu_1-\mu_2)}{\displaystyle \sqrt {(\frac {SS_1+SS_2}{n_1+n_2-2})(\frac {1}{n_1}+\frac {1}{n_2})}}\end{equation*}$<br>

從上面的內容可以知道，當兩個樣本來自的母群體的**變異數無法得知**時，可以使用t分配來描述樣本平均數差異的抽樣分配，並運用

 $\begin{equation*}t=\frac {(\overline X_1-\overline X_2)-(\mu_1-\mu_2)}{\displaystyle \sqrt {(\frac {SS_1+SS_2}{n_1+n_2-2})(\frac {1}{n_1}+\frac {1}{n_2})}}\end{equation*}$<br>
 
來計算t檢定統計量，以便進行獨立樣本t檢定的假設檢定。

**獨立樣本t檢定的假設檢定**

獨立樣本t檢定用來比較兩個樣本平均數是否有顯著的不同，且在樣本來自的**母群體變異數$\sigma^2$未知**的情況下使用。<br>
換句話說，該種統計檢定方法通常評估$\mu_1=\mu_2$的虛無假設（$H_0$），也就是兩個樣本是否來自於兩個母群體平均數相同的一組隨機樣本。<br>

如同獨立樣本z檢定:

研究假設:<br>
可分為對立假設（$H_1$）和虛無假設（$H_0$）<br>
依據方向:<br>
劃分為無方向性和有方向性的假設<br>

- 無方向性的研究假設<br>
  - 對立假設（$H_1$）：第1個樣本的平均數不同於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 \neq \mu_2$的隨機樣本。<br>
  - 虛無假設（$H_0$）：第1個樣本的平均數和第2個樣本的平均數相同，也就是樣本是來自於母群體$\mu_1 = \mu_2$的隨機樣本。<br>


- 有方向性的研究假設<br>

  若明確地指出自變項效果或作用的方向，則為有方向性的研究假設，可能會有$\mu_1 > \mu_2$或$\mu_1 < \mu_2$的兩種情況，須分開來看。<br>

  - 若$\mu_1 > \mu_2$<br>
    - 對立假設（$H_1$）：第1個樣本的平均數大於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 > \mu_2$的隨機樣本。<br>
    - 虛無假設（$H_0$）：第1個樣本的平均數沒有大於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 \leq \mu_2$的隨機樣本。<br>
 
  - 若$\mu_1 < \mu_2$<br>
    - 對立假設（$H_1$）：第1個樣本的平均數小於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 < \mu_2$的隨機樣本。<br>
    - 虛無假設（$H_0$）：第1個樣本的平均數沒有小於第2個樣本的平均數，也就是樣本是來自於母群體$\mu_1 \geq \mu_2$的隨機樣本。<br>
    
獨立樣本t檢定使用t分配和t檢定統計量，而<br>
**t檢定統計量的公式:**<br>

$\begin{equation*}t=\frac {(\overline X_1-\overline X_2)-(\mu_1-\mu_2)}{\displaystyle \sqrt {(\frac {SS_1+SS_2}{n_1+n_2-2})(\frac {1}{n_1}+\frac {1}{n_2})}}\end{equation*}$<br>

因為在多數的情況下，我們是評估$\mu_1=\mu_2$的虛無假設，也就是$\mu_1-\mu_2=0$，所以可再稍微簡化成下面的公式：

$\begin{equation*}t=\frac {\overline X_1-\overline X_2}{\displaystyle \sqrt {(\frac {SS_1+SS_2}{n_1+n_2-2})(\frac {1}{n_1}+\frac {1}{n_2})}}\end{equation*}$

**決策規則:**

- 獲得t檢定統計量後，**查詢t分配表**，**自由度為$n_1+n_2-2=N-2$（N為兩個樣本的總和）**，依循事先選擇好的$\alpha$水準，找到相對應的臨界值。<br>
  比較t檢定統計量和臨界值，當**t檢定統計量的絕對值等於或大於臨界值的絕對值**時，即可拒絕虛無假設。
  

- 獲得t檢定統計量的機率（p值）而不是臨界值，這時須使用機率比較的決策規則，當**p值小於或等於$\alpha$水準時（$p \leq \alpha$）**，即可拒絕虛無假設，接受對立假設；反之，則須保留虛無假設。
  
若是有方向性的研究假設，須注意**t檢定統計量的符號須和研究假設的方向一致**，否則就應保留虛無假設。<br>
例如:<br>
對立假設為第1個樣本的平均數小於第2個樣本的平均數，計算出來的t檢定統計量應為負數（$\overline X_1 < \overline X_2 \Rightarrow \overline X_1-\overline X_2 < 0$），才能進一步比較t檢定統計量和臨界值；<br>
相反地，若t檢定統計量為正數，因為和對立假設的方向不符，就應保留虛無假設。

**獨立樣本t檢定的基本假設**

獨立樣本t檢定是個很常被使用的統計檢定方法，<br>
是一種**對樣本來自的母群體做出特定假設的母數檢定（parametric test）**。<br>

當利用獨立樣本t檢定進行假設檢定時，樣本資料須滿足下面的兩個假設：<br>

1. **常態分配**：樣本平均數差異（$\overline X_1-\overline X_2$）的**抽樣分配須呈現常態分配**，這也是指樣本來自的母群體必須是常態分配的型態。
2. **變異數同質性（homogeneity of variance）**：**兩個樣本來自的母群體的變異是相同的，也就是$\sigma_1^2=\sigma_2^2$**的情況。如果兩個樣本一開始就存在很大的變異，很有可能不是來自於母群體$\sigma_1^2=\sigma_2^2$的隨機樣本，即是違反了該假設。<br>
   (第1個母群體的變異和第2個母群體的**變異相同**，也就是$\sigma_1^2 = \sigma_2^2$，稱為**變異數同質性（homogeneity of variance）假設**。<br>
  當自變項確實有效果或作用時，該效果會影響到群體裡的每一個人，因此**只會改變母群體的平均數，但不會改變母群體的標準差或變異**。)<br>
<br>

雖然執行獨立樣本t檢定時，樣本資料最好能夠滿足上述的兩個假設，<br>
但獨立樣本t檢定可說是一種穩健（robust）的檢定方法，即使違反了假設，該檢定的執行結果仍會如預期（Sawilosky & Blair, 1992）。<br>
更正確地說，**如果兩個樣本的樣本大小一致（$n_1=n_2$）、每一個樣本的樣本大小$\geq 25$且使用雙尾檢定，即使輕度或中度違反假設，檢定的結果仍具有可信度。**

然而，**若兩個樣本的樣本大小不一樣，解釋研究結果時就要特別小心**。<br>
如果兩個樣本的樣本大小差異很大或變異很大，獨立樣本t檢定可能就不是一個合適的統計檢定方法，<br>
此時可考慮其他的檢定方法，例如調整不相等變異的Welch´s t檢定，或無母數檢定裡的曼‐惠特尼 U檢定（Mann-Whitney U test）。<br>

### **關聯樣本t檢定的假設檢定**

關聯樣本t檢定（correlated samples t-test）用來比較**兩個"相關聯"的樣本或群組的平均數**是否有顯著的差異，也稱為成對樣本t檢定（paired samples t-test）。<br>
這種統計檢定方法通常用在**重複量數設計（repeated-measures design）**或**配對組設計（matched-group design）**的研究設計上，<br>
是實驗設計裡很常使用到的統計檢定方法之一。<br>

由於兩樣本之間具有關聯性，<br>
所以計算檢定統計量時不是使用原始分數（raw scores），<br>
而是**配對的差值分數**來進行分析，又稱為直接差分法（direct difference method）。<br>
這種方法將原本屬於兩組樣本的分析轉換成單一樣本的分析，<br>
換句話說，**關聯樣本t檢定可看成是平均數差值的單一樣本t檢定**<br>

**關聯樣本的研究設計**

關聯樣本（或稱為關聯群組、相依群組）的研究設計是指**樣本之間或群組之間並非各自獨立，而是具有關聯性**，<br>
也**不用隨機分配研究參與者（或稱受試者）至實驗裡的不同情境**。<br>
這種研究設計通常可分為兩種類型，一種是重複量數設計，另一種則是配對組設計。<br>

- 重複量數設計<br>
  重複量數設計有時也稱為受試者內設計（within-subjects design）或前後測設計（before-after design），<br>
  是指**相同的受試者參與兩種或超過兩種的實驗情境**。<br>
  在這種研究設計裡，因為受試者本身即是實驗過程中的控制條件，所以可降低個人差異所導致之誤差，反而更能**偵測到實驗裡自變項的效果**。<br>
  
  例子:<br>
  舉例來說，若有位社會工作者想探討親密伴侶暴力防治宣導課程的效果，她在**課程開始前**請參與的學員填寫一份與課程內容相關的知識問卷，<br>
  **課程結束後**再請學員們填寫一次問卷。<br>
  這位社工應可預期每位學員的表現相對地前後一致，也就是前測成績不錯的學員，後測的成績也會不錯，<br>
  而且大多數的學員成績在宣導課程結束後可能都有進步，使得前測和後測的成績間具有關聯性。<br>
  
  這個因為**相同學員的參與所導致的前後測成績的關聯性是可以被測量出來的**，<br>
  所以在統計分析時可以移除該變異（也就是關聯性），增加了宣導課程確實具備效果的可能性。<br>
  
  不過重複量數設計有一個**很大的缺點**，就是**轉移效應（carryover effect）**。<br>
  轉移效應是指**第一次測驗影響到第二次測驗的結果**，<br>
  尤其是當研究人員使用標準化測驗的時候，受試者可能會記得問題和答案，進而使得第二次測驗結果有所偏誤。<br>
  **若轉移效應確實會嚴重影響到研究的信度，下面介紹的配對組設計可能會是較合適的研究設計。**<br>
  

- 配對組設計<br>
  配對組設計指研究人員藉由和**依變項相關的其他變項的測量**，<br>
  將**測量結果類似的受試者進行配對，再隨機分配每一配對裡的其中一人至一個實驗情境、配對裡的另一人至另一個實驗情境。**<br>
  
  例子:<br>
  舉例來說，若社會工作者相信學員的生理性別、教育程度和之前參與類似宣導課程的次數會影響親密伴侶暴力防治宣導課程的效果，<br>
  因此利用這3個變項將學員進行配對，**具備類似特質者配成一對**，<br>
  例如生理女性、高中教育程度和從未參與過類似課程的2位學員為一對，生理男性、大學教育程度和參與過1次類似課程的2位學員為另一對。<br>
  配對完成後，社會工作者將每一配對的其中一人隨機分配至親密伴侶暴力防治宣導課程組，配對裡的另一人則至通識教育課程組。<br>
  課程結束後，每人皆填寫一份親密伴侶暴力防治的相關問卷，再比較兩組的成績是否有顯著的不同。<br>
  
  配對組設計的優點是**確保兩組具有類似的特質**，<br>
  而且和重複量數設計一樣，能夠測量配對受試者之間的關聯性並在統計分析時移除該變異。<br>
  另外，這種研究設計**不會有轉移效應**的問題，是個**蠻常被使用的關聯樣本的研究設計**。<br>
  
  
**關聯樣本t檢定的檢定統計量公式**

依據變異數總和法則（variance sum law），在兩個獨立樣本的設計裡，樣本平均數差異抽樣分配的變異等於兩樣本來自的母群體變異數之總和：

$ \sigma_{\overline X_1-\overline X_2}^2=\sigma_{\overline X_1}^2+\sigma_{\overline X_2}^2 $

當母群體變異數未知的時候，會利用樣本變異數$s^2$來估計母群體變異數，則樣本平均數差異的標準誤估計值如下：

$\begin{equation*}s_{\overline X_1-\overline X_2}=\sqrt {s_{\overline X_1}^2+s_{\overline X_2}^2}\end{equation*}$

不過，實際上**樣本平均數差異的標準誤估計值**的通式為下面的方程式：

$\begin{equation*}s_{\overline X_1-\overline X_2}=\sqrt {s_{\overline X_1}^2+s_{\overline X_2}^2-2r(s_{\overline X_1})(s_{\overline X_2})}\end{equation*}$

$r$指兩個樣本或群組間的關聯性。<br>
在獨立樣本的設計裡，因為兩個樣本各自獨立，沒有關聯性，所以$r=0$，<br>
使得$2r(s_{\overline X_1})(s_{\overline X_2})=0$，樣本平均數差異標準誤的估計值才會是上面公式的樣子。<br>

概念上**關聯樣本t檢定的檢定統計量公式**為：

$\begin{equation*}t=\frac {(\overline X_1-\overline X_2)-(\mu_1-\mu_2)}{\sqrt {s_{\overline X_1}^2+s_{\overline X_2}^2-2r(s_{\overline X_1})(s_{\overline X_2})}}\end{equation*}$

不過利用此公式來進行運算，過程會有點複雜，<br>
所以還有一種計算上較友善且能夠得到和一樣結果的方法稱為**直接差分法**，<br>
這種方法**利用各個配對分數的差值進行運算，將原本屬於兩個樣本的檢定變成單一樣本的檢定。**<br>

**關聯樣本t檢定的檢定統計量**為：

$\begin{equation*}t=\frac {\overline D-\mu_D}{\displaystyle \frac {s_D}{\sqrt N}}\end{equation*}$

$D$代表配對的差值分數<br>
$\overline D$為樣本差值分數的平均數<br>
$\mu_D$為母群體差值分數的平均數<br>
$s_D$為樣本差值分數的標準差<br>
N為差值分數的總個數<br>

樣本差值分數的平均數實際上為兩個樣本平均數的差值，也就是$\overline D=\overline X_1-\overline X_2$

=> $\begin{equation*}t=\frac {\overline D-\mu_D}{\displaystyle \frac {s_D}{\sqrt N}}=\frac {\overline D-\mu_D}{\sqrt {\displaystyle \frac {\sum D^2-\displaystyle \frac {(\sum D)^2}{N}}{N(N-1)}}}\end{equation*}$

兩者的差別在於單一樣本t檢定使用**原始分數**，而關聯樣本t檢定使用**配對的差值分數**來進行分析。

**關聯樣本t檢定的假設檢定**

關聯樣本t檢定用來探討兩個相關聯的樣本或群組的平均數是否有顯著的不同，<br>
更簡單地說，是在**評估自變項是否帶有效果**。<br>
若用上面提到的**配對差值分數**的概念來看，<br>
這種檢定方法用在評估母群體差值分數平均數$\mu_D=0$的虛無假設，又因為$\mu_D=\mu_1-\mu_2$，所以也可說是評估$\mu_1-\mu_2=0$的虛無假設（$H_0$）。<br>

關聯樣本t檢定使用t分配和t檢定統計量，其中t檢定統計量的計算公式即為:

$\begin{equation*}t=\frac {\overline D-\mu_D}{\displaystyle \frac {s_D}{\sqrt N}}=\frac {\overline D-\mu_D}{\sqrt {\displaystyle \frac {\sum D^2-\displaystyle \frac {(\sum D)^2}{N}}{N(N-1)}}}\end{equation*}$

由於在大多數的情況下，研究人員是評估$\mu_D=\mu_1-\mu_2=0$的虛無假設，<br>
所以公式可再簡化為下面的公式：<br>
$\begin{equation*}t=\frac {\overline D}{\sqrt {\displaystyle \frac {\sum D^2-\displaystyle \frac {(\sum D)^2}{N}}{N(N-1)}}}\end{equation*}$<br>

**決策規則:**

- 獲得t檢定統計量後，**查詢t分配表**，**自由度為$n_1+n_2-2=N-2$（N為兩個樣本的總和）**，依循事先選擇好的$\alpha$水準，找到相對應的臨界值。<br>
  比較t檢定統計量和臨界值，當**t檢定統計量的絕對值等於或大於臨界值的絕對值**時，即可拒絕虛無假設。
  

- 獲得t檢定統計量的機率（p值）而不是臨界值，這時須使用機率比較的決策規則，當**p值小於或等於$\alpha$水準時（$p \leq \alpha$）**，即可拒絕虛無假設，接受對立假設；反之，則須保留虛無假設。
  
若是有方向性的研究假設，須注意**t檢定統計量的符號須和研究假設的方向一致**，否則就應保留虛無假設。<br>

關聯樣本t檢定和其他的統計檢定方法一樣，樣本資料須滿足特定的假設後才可使用。

**關聯樣本t檢定的基本假設**

關聯樣本t檢定是對**樣本來自的母群體**做出特定假設的母數檢定（parametric test），<br>
所以執行假設檢定時，樣本資料須滿足特定的條件。<br>
因為**關聯樣本t檢定類似於單一樣本t檢定**，所以樣本資料須符合的假設也和單一樣本t檢定相同。<br>

平均數抽樣分配須呈現常態分配，也就是樣本數等於或大於30，或樣本來自的母群體呈現常態分配。<br>
簡單地說，關聯樣本t檢定要求**樣本配對差值分數平均數$\overline D$的抽樣分配須呈現常態分配，**<br>
為了達到該要求，**配對差值分數的總個數$N \geq 30$**，或樣本來自的母群體分數本身即為常態分配。<br>

### **獨立樣本和關聯樣本t檢定的比較**

- 獨立樣本t檢定:<br>
  比較**兩個獨立群體的平均數**是否有顯著的不同。<br>
  研究設計:使用**不同受試者**的**獨立樣本設計**。<br>

- 關聯樣本t檢定（也稱為成對樣本t檢定）:<br>
  比較**兩個相關聯群體的平均數**是否有顯著的差異。<br>
  研究設計:使用**相同或配對受試者**的**重複量數**或**配對組設計**。<br>


雖然這兩種統計檢定方法的使用時機分明，但為什麼要使用重複量數或配對組的研究設計而不直接使用獨立樣本的設計就好呢？<br>
原因在於關聯樣本設計的**變異（誤差）小於獨立樣本的設計**，<br>
使得關聯樣本t檢定的t檢定統計量t比獨立樣本t檢定的檢定統計量更為強力（powerful），因此更容易拒絕虛無假設<br>

- 獨立樣本t檢定的檢定統計量:

  $\begin{equation*}t=\frac {(\overline X_1-\overline X_2)-(\mu_1-\mu_2)}{\displaystyle \sqrt {(\frac {SS_1+SS_2}{n_1+n_2-2})(\frac {1}{n_1}+\frac {1}{n_2})}}\end{equation*}$<br>

  因為在多數的情況下，我們是評估$\mu_1=\mu_2$的虛無假設，也就是$\mu_1-\mu_2=0$，所以可再稍微簡化成下面的公式：

  $\begin{equation*}t=\frac {\overline X_1-\overline X_2}{\displaystyle \sqrt {(\frac {SS_1+SS_2}{n_1+n_2-2})(\frac {1}{n_1}+\frac {1}{n_2})}}\end{equation*}$


- 關聯樣本t檢定的檢定統計量:

  **樣本平均數差異的標準誤估計值**的通式為下面的方程式：

  $\begin{equation*}s_{\overline X_1-\overline X_2}=\sqrt {s_{\overline X_1}^2+s_{\overline X_2}^2-2r(s_{\overline X_1})(s_{\overline X_2})}\end{equation*}$

  $r$指兩個樣本或群組間的關聯性。<br>
  在獨立樣本的設計裡，因為兩個樣本各自獨立，沒有關聯性，所以$r=0$，<br>
  使得$2r(s_{\overline X_1})(s_{\overline X_2})=0$，樣本平均數差異標準誤的估計值才會是上面公式的樣子。<br>

  概念上**關聯樣本t檢定的檢定統計量公式**為：

  $\begin{equation*}t=\frac {(\overline X_1-\overline X_2)-(\mu_1-\mu_2)}{\sqrt {s_{\overline X_1}^2+s_{\overline X_2}^2-2r(s_{\overline X_1})(s_{\overline X_2})}}\end{equation*}$

  **直接差分法**:<br>
  這種方法**利用各個配對分數的差值進行運算，將原本屬於兩個樣本的檢定變成單一樣本的檢定。**<br>

  **關聯樣本t檢定的檢定統計量**為：

  $\begin{equation*}t=\frac {\overline D-\mu_D}{\displaystyle \frac {s_D}{\sqrt N}}=\frac {\overline D-\mu_D}{\sqrt {\displaystyle \frac {\sum D^2-\displaystyle \frac {(\sum D)^2}{N}}{N(N-1)}}}\end{equation*}$

  $D$代表配對的差值分數<br>
  $\overline D$為樣本差值分數的平均數<br>
  $\mu_D$為母群體差值分數的平均數<br>
  $s_D$為樣本差值分數的標準差<br>
  N為差值分數的總個數<br>

  因為在多數的情況下，我們是評估$\mu_1=\mu_2$的虛無假設，也就是$\mu_1-\mu_2=0$，所以可再稍微簡化成下面的公式：

  $\begin{equation*}t=\frac {\overline D}{\sqrt {\displaystyle \frac {\sum D^2-\displaystyle \frac {(\sum D)^2}{N}}{N(N-1)}}}\end{equation*}$<br>



關聯樣本的設計使用**受試者本身作為研究的控制因子**，<br>
排除了因為受試者個人的因素所致的變異，進而**降低了整體的變異（誤差）**，<br>
使得t檢定統計量的數值變大，因此**更容易拒絕虛無假設。**<br>

雖然**關聯樣本t檢定統計量較獨立樣本t檢定統計量更為強力**，<br>
但關聯樣本的研究設計具有一些固有的限制，並非隨時隨地都可使用，以下來探討這些限制：

- 自由度少於獨立樣本的設計：<br>
  自由度是指計算一個統計量時，可以自由變動的分數個數，是影響統計分析結果的一個重要因素。<br>
  **當自由度愈高的時候，臨界值愈小，愈有可能拒絕虛無假設。**<br>
  從上面的例子來看，在關聯樣本的設計裡，雖然有20筆資料（每一位小學生都有2次測量），但自由度只有9；<br>
  而在獨立樣本的設計裡，同樣有20筆資料，自由度卻為18。因此，**從自由度的角度來看，獨立樣本設計比關聯樣本設計更具有效性。**<br>


- 並非所有的研究都可使用相同的受試者：<br>
  有些研究很明顯地無法在兩個不同的情境裡使用相同的受試者或研究參與者，<br>
  例如涉及生理男性和女性、懷孕者和沒有懷孕者、受虐者和非受虐者等研究。<br>
  另外，若第1個情境的作用會影響到第2個情境，也不適合使用相同的受試者，<br>
  例如探討某種讀書技巧的效果，當受試者已經在第1個情境中習得讀書技巧，他們就不適合再作為第2個情境的受試者。<br>


- 配對耗費時間和資源：<br>
  雖然在無法使用相同受試者的情況下，**可改採用配對的受試者**，<br>
  但為了瞭解和取得配對所需的資訊，通常須耗費難以估算的時間、邏輯和金錢。<br>
  此外，退出率高也會影響配對組設計，因為當其中一人退出的時候，配對的另一人也無法繼續留在研究裡。<br>
  最後，配對組設計可能會選擇性地捨棄一些特殊或極端的受試者，因為無法找到合適的配對者，反而會造成樣本無法反映出真實母群體的情況。<br>
  
因此，固然**關聯樣本t檢定較獨立樣本t檢定靈敏**，<br>
運用關聯樣本的設計有時也能夠以成本效益的方式獲得以往難以發現的研究成果，<br>
但基於上面幾點關聯樣本的固有限制，使得**獨立樣本設計仍舊是較常被使用的研究設計。**<br>


但不論如何，研究人員都應依據研究的方向和目的，在理論基礎上，選擇合適的研究設計和統計分析方法，使研究結果得以反映真實的情況。
